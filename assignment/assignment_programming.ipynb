Now do it again, with slightly less direction:

8. Create a grid of 100 equally spaced points ranging from -6.5 to 6.5.
9. Use the sine and cosine functions in Numpy to compute the values of those functions for each point on your grid. (You'll have to find out what those functions are.)

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 1, 50)


y = np.log(x)

z = np.exp(x)


plt.scatter(x,y,color="blue",label='Natural Log')
plt.xlabel("Number")
plt.ylabel("Natural Log")
plt.title("Number vs. Natural Log (of 50 numbers between 0 and 1)")
plt.legend(loc = 'upper left')
plt.show()



plt.scatter(x,z,color="blue",label='Exponential')
plt.xlabel("Number")
plt.ylabel("Exponential")
plt.title("Number vs. Exponential (of 50 numbers between 0 and 1)")
plt.legend(loc = 'lower right')
plt.show()


```

```python

import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-6.5, 6.5, 100)

y = np.sin(x)

z = np.cos(x)



plt.plot(y,z,color="blue",label='Numbers')
plt.xlabel("Numbers")
plt.ylabel("Cosine")
plt.title("Sine vs. Cosine (of 100 numbers between -6.5 and 6.5)")
plt.legend(loc = 'lower left')
plt.show()
```
